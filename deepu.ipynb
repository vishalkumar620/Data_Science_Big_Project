```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
import numpy as np

# Generate synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Instantiate models
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
knn_classifier = KNeighborsClassifier()
dt_classifier = DecisionTreeClassifier(random_state=42)
linear_reg_model = LinearRegression()

# Train the models
rf_classifier.fit(X_train, y_train)
knn_classifier.fit(X_train, y_train)
dt_classifier.fit(X_train, y_train)
linear_reg_model.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_classifier.predict(X_test)
y_pred_knn = knn_classifier.predict(X_test)
y_pred_dt = dt_classifier.predict(X_test)
y_pred_linear = linear_reg_model.predict(X_test)

# Evaluate the models
accuracy_rf = accuracy_score(y_test, y_pred_rf)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
accuracy_linear = accuracy_score((y_test >= np.median(y)), (y_pred_linear >= 0.5))

# Print accuracies
print("Random Forest Accuracy:", accuracy_rf)
print("K-Nearest Neighbors Accuracy:", accuracy_knn)
print("Decision Tree Accuracy:", accuracy_dt)
print("Linear Regression Accuracy:", accuracy_linear)
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import pandas as pd

# Generate synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)

# Binarize the target variable for classification
y_binary = np.where(y >= np.median(y), 1, 0)

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=42)

# Instantiate models
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
knn_classifier = KNeighborsClassifier()
dt_classifier = DecisionTreeClassifier(random_state=42)
linear_reg_model = LinearRegression()

# Train the models
rf_classifier.fit(X_train, y_train)
knn_classifier.fit(X_train, y_train)
dt_classifier.fit(X_train, y_train)
linear_reg_model.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_classifier.predict(X_test)
y_pred_knn = knn_classifier.predict(X_test)
y_pred_dt = dt_classifier.predict(X_test)
y_pred_linear = np.where(linear_reg_model.predict(X_test) >= 0.5, 1, 0)

# Evaluate the models
accuracy_rf = accuracy_score(y_test, y_pred_rf)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
accuracy_linear = accuracy_score(y_test, y_pred_linear)

# Confusion matrices
confusion_rf = confusion_matrix(y_test, y_pred_rf)
confusion_knn = confusion_matrix(y_test, y_pred_knn)
confusion_dt = confusion_matrix(y_test, y_pred_dt)
confusion_linear = confusion_matrix(y_test, y_pred_linear)

# Create Pandas DataFrames for confusion matrices
df_rf = pd.DataFrame(confusion_rf, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])
df_knn = pd.DataFrame(confusion_knn, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])
df_dt = pd.DataFrame(confusion_dt, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])
df_linear = pd.DataFrame(confusion_linear, index=['Actual 0', 'Actual 1'], columns=['Predicted 0', 'Predicted 1'])

# Print accuracies
print("Random Forest Accuracy:", accuracy_rf)
print("K-Nearest Neighbors Accuracy:", accuracy_knn)
print("Decision Tree Accuracy:", accuracy_dt)
print("Linear Regression Accuracy:", accuracy_linear)

# Print confusion matrices
print("\nConfusion Matrix for Random Forest:")
print(df_rf)

print("\nConfusion Matrix for K-Nearest Neighbors:")
print(df_knn)

print("\nConfusion Matrix for Decision Tree:")
print(df_dt)

print("\nConfusion Matrix for Linear Regression:")
print(df_linear)
```

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate synthetic dataset
X, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)

# Binarize the target variable for classification
y_binary = np.where(y >= np.median(y), 1, 0)

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_binary, test_size=0.3, random_state=42)

# Instantiate models
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
knn_classifier = KNeighborsClassifier()
dt_classifier = DecisionTreeClassifier(random_state=42)
linear_reg_model = LinearRegression()

# Train the models
rf_classifier.fit(X_train, y_train)
knn_classifier.fit(X_train, y_train)
dt_classifier.fit(X_train, y_train)
linear_reg_model.fit(X_train, y_train)

# Make predictions
y_pred_rf = rf_classifier.predict(X_test)
y_pred_knn = knn_classifier.predict(X_test)
y_pred_dt = dt_classifier.predict(X_test)
y_pred_linear = np.where(linear_reg_model.predict(X_test) >= 0.5, 1, 0)

# Evaluate the models
accuracy_rf = accuracy_score(y_test, y_pred_rf)
accuracy_knn = accuracy_score(y_test, y_pred_knn)
accuracy_dt = accuracy_score(y_test, y_pred_dt)
accuracy_linear = accuracy_score(y_test, y_pred_linear)

# Confusion matrices
confusion_rf = confusion_matrix(y_test, y_pred_rf)
confusion_knn = confusion_matrix(y_test, y_pred_knn)
confusion_dt = confusion_matrix(y_test, y_pred_dt)
confusion_linear = confusion_matrix(y_test, y_pred_linear)

# Plot confusion matrices
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
sns.heatmap(confusion_rf, annot=True, fmt='d', cmap='Blues', cbar=False, annot_kws={"size": 12})
plt.title('Random Forest Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')

plt.subplot(2, 2, 2)
sns.heatmap(confusion_knn, annot=True, fmt='d', cmap='Blues', cbar=False, annot_kws={"size": 12})
plt.title('K-Nearest Neighbors Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')

plt.subplot(2, 2, 3)
sns.heatmap(confusion_dt, annot=True, fmt='d', cmap='Blues', cbar=False, annot_kws={"size": 12})
plt.title('Decision Tree Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')

plt.subplot(2, 2, 4)
sns.heatmap(confusion_linear, annot=True, fmt='d', cmap='Blues', cbar=False, annot_kws={"size": 12})
plt.title('Linear Regression Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')

plt.tight_layout()
plt.show()
```
